```python
import matplotlib.pyplot as plt
import ternary
from ternary.helpers import simplex_iterator
```

```python
def color_point(x, y, z, scale):
    r = x / scale
    g = y / scale
    b = z / scale
    return (r, g, b, 1)

def generate_heatmap(scale=5):
    d = dict()
    for (i,j,k) in simplex_iterator(scale):
        d[(i,j,k)] = color_point(i,j,k,scale)
    return d

scale = 10
data = generate_heatmap(scale)
figure, tax = ternary.figure(scale=scale)
tax.heatmap(data, style='triangular', use_rgba=True)
tax.boundary(linewidth=2)
tax.ticks(axis='lrb', offset=0.03)
plt.show()
```

first value is max. at bottom right, second value is max. at top, third value is max at bottom left.

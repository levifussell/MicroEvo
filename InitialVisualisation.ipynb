```python
def process_and_plot(params, ax):
    name = "data/k{}_i{}_d{}_m{}_ep{}.csv".format(
            params['k'],
            params['i'],
            params['d'],
            params['m'],
            params['ep'])
    data = np.loadtxt(name, delimiter=',', dtype=str)
    data = data[:, :-1].astype(int).transpose()
    data = data[np.where(np.sum(data, 1) != 0)]
#     fig, ax = plt.subplots(figsize=(10,7))
    for d in range(data.shape[0]):
        ax.plot(range(data.shape[1]), data[d, :])
    ax.set_title(name)

def process_param(params):
    total = 1
    for k in params.keys():
        total *= len(params[k])
    fig, ax = plt.subplots(total, 1, figsize=(10,7*total))
    idx = 0
    for k in params['k']:
        for i in params['i']:
            for d in params['d']:
                for m in params['m']:
                    for ep in params['ep']:
                        process_and_plot({
                            'k':k,
                            'i':i,
                            'd':d,
                            'm':m,
                            'ep':ep,
                        }, ax[idx])
                        idx += 1
```

```python
process_param({'k':[3], 'i':[0], 'd':[3], 'm':[0.02], 'ep':[0.1, 0.2, 0.3]})
```
